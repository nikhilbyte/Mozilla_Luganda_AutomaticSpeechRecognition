```python
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
from transformers import Wav2Vec2CTCTokenizer

tokenizer = Wav2Vec2Tokenizer.from_pretrained("lucio/wav2vec2-large-xlsr-luganda")
```

```python
model = Wav2Vec2ForCTC.from_pretrained("/content/drive/MyDrive/LUGANDA_ASR_MODEL/checkpoint-37200",vocab_size = 32)
```

```python
import os
```

```python
results = dict()
for files in os.listdir("/content/test_audio"):
  speech, rate = librosa.load(os.path.join("/content/test_audio",files),sr=16000)
  input_values = tokenizer(speech, return_tensors = 'pt').input_values
  logits = model(input_values).logits
  predicted_ids = torch.argmax(logits, dim =-1)
  transcriptions = tokenizer.decode(predicted_ids[0])
  results[files] = transcriptions
```

```python
results
```

```python
ss["Clip_ID"] = ss["Clip_ID"]+".mp3"
```

```python
ss["Target"] = ss["Clip_ID"].map(results)
```
